```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('archive/airports.csv')

# Data Cleaning and Exploration
# Let's check the first few rows of the dataset
print(data.head())

# Check for missing values
print(data.isnull().sum())

# Summary statistics
print(data.describe())

# Data Visualization
# Create a line plot of Latitude vs. Longitude
plt.figure(figsize=(10, 6))
plt.plot(data['LONGITUDE'], data['LATITUDE'], 'bo', markersize=3)
plt.title('Airport Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.show()

# Create a scatter plot of Airport Count by State
state_counts = data['STATE'].value_counts()
plt.figure(figsize=(12, 6))
plt.scatter(state_counts.index, state_counts.values, marker='o', s=50)
plt.title('Airport Count by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
